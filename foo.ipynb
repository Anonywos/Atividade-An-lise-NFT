```python
import networkx as nx
import numpy as np
import itertools

# Cria um grafo direcionado
G = nx.DiGraph()

# Adiciona arestas ao grafo com pesos baseados no valor da transação
for index, row in df_new.iterrows():
	G.add_edge(row['Seller_address'], row['Buyer_address'], weight=row['Price_USD'])
```

```python
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Is directed:", G.is_directed())

def analyze_degrees(G):
	print("--- Análise de Graus dos Vértices ---")
	if G.is_directed():
		degrees = dict(G.degree())
		in_degrees = dict(G.in_degree())
		out_degrees = dict(G.out_degree())
		print(f"Grau total: Média={np.mean(list(degrees.values())):.2f}, "
			f"Mediana={np.median(list(degrees.values())):.2f}")
		print(f"Grau de entrada: Média={np.mean(list(in_degrees.values())):.2f}, "
			f"Mediana={np.median(list(in_degrees.values())):.2f}")
		print(f"Grau de saída: Média={np.mean(list(out_degrees.values())):.2f}, "
			f"Mediana={np.median(list(out_degrees.values())):.2f}")
	else:
		degrees = dict(G.degree())
		print(f"Grau total: Média={np.mean(list(degrees.values())):.2f}, "
			f"Mediana={np.median(list(degrees.values())):.2f}")

def analyze_shortest_paths(G):
	print("\n--- Análise de Caminhos Mínimos ---")
	try:
		path_lengths = list(itertools.chain.from_iterable(
			dict(nx.all_pairs_shortest_path_length(G)).values()
		))
		print(f"Média do tamanho dos caminhos mínimos: {np.mean(path_lengths):.2f}")
		print(f"Mediana do tamanho dos caminhos mínimos: {np.median(path_lengths):.2f}")
		print(f"Desvio padrão: {np.std(path_lengths):.2f}")
	except nx.NetworkXError:
		print("O grafo é desconexo, não é possível calcular todos os caminhos mínimos.")

# def analyze_components(G):
# 	print("\n--- Análise de Componentes Conectados ---")
# 	if G.is_directed():
# 		components = list(nx.weakly_connected_components(G))
# 	else:
# 		components = list(nx.connected_components(G))

# 	print(f"Quantidade de componentes conectados: {len(components)}")
# 	largest_component = max(components, key=len)
# 	print(f"Tamanho do maior componente: {len(largest_component)}")

def analyze_components(G):
    print("\n--- Análise de Componentes Conectados ---")
    
    # Verifica se o grafo é direcionado ou não
    if G.is_directed():
        components = list(nx.weakly_connected_components(G))  # Componentes fracos
    else:
        components = list(nx.connected_components(G))  # Componentes fortes
    
    # Quantidade de componentes conectados
    num_components = len(components)
    print(f"Quantidade de componentes conectados: {num_components}")
    
    # Tamanho do maior componente
    largest_component = max(components, key=len)
    print(f"Tamanho do maior componente: {len(largest_component)}")
    
    # Distribuição de tamanhos dos componentes
    component_sizes = [len(component) for component in components]
    
    # Média e desvio padrão dos tamanhos dos componentes
    mean_size = np.mean(component_sizes)
    std_size = np.std(component_sizes)
    
    print(f"Média do tamanho dos componentes: {mean_size:.2f}")
    print(f"Desvio padrão do tamanho dos componentes: {std_size:.2f}")
    
    # Número de componentes pequenos (tamanho <= 10)
    small_components = sum(1 for size in component_sizes if size <= 10)
    print(f"Quantidade de componentes pequenos (tamanho <= 10): {small_components}")
    
    # Número de componentes grandes (tamanho > 100)
    large_components = sum(1 for size in component_sizes if size > 100)
    print(f"Quantidade de componentes grandes (tamanho > 100): {large_components}")
    
    # Histograma de tamanhos dos componentes
    print("\nHistograma dos tamanhos dos componentes:")
    unique_sizes, counts = np.unique(component_sizes, return_counts=True)
    for size, count in zip(unique_sizes, counts):
        print(f"Tamanho {size}: {count} componentes")
print("Análise")
analyze_components(G)
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def analyze_graph_with_visuals(G):
    # Calcula os graus de entrada, saída e total
    in_degrees = dict(G.in_degree())  # Grau de entrada
    out_degrees = dict(G.out_degree())  # Grau de saída
    total_degrees = {node: in_degrees[node] + out_degrees[node] for node in G.nodes()}  # Grau total

    # Converte para listas
    in_degree_values = list(in_degrees.values())
    out_degree_values = list(out_degrees.values())
    total_degree_values = list(total_degrees.values())

    # Configuração dos gráficos
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Histograma do Grau Total
    axes[0, 0].hist(total_degree_values, bins=50, color='blue', alpha=0.7)
    axes[0, 0].set_title("Distribuição do Grau Total (Compras + Vendas)")
    axes[0, 0].set_xlabel("Número de Conexões")
    axes[0, 0].set_ylabel("Frequência")
    axes[0, 0].set_yscale("log")  # Escala logarítmica para melhor visualização

    # Histograma do Grau de Entrada
    axes[0, 1].hist(in_degree_values, bins=50, color='green', alpha=0.7)
    axes[0, 1].set_title("Distribuição do Grau de Entrada (Vendas)")
    axes[0, 1].set_xlabel("Número de Vendas")
    axes[0, 1].set_ylabel("Frequência")
    axes[0, 1].set_yscale("log")

    # Histograma do Grau de Saída
    axes[1, 0].hist(out_degree_values, bins=50, color='red', alpha=0.7)
    axes[1, 0].set_title("Distribuição do Grau de Saída (Compras)")
    axes[1, 0].set_xlabel("Número de Compras")
    axes[1, 0].set_ylabel("Frequência")
    axes[1, 0].set_yscale("log")

    # Boxplot dos Graus
    axes[1, 1].boxplot([total_degree_values, in_degree_values, out_degree_values], labels=['Grau Total', 'Grau Entrada', 'Grau Saída'])
    axes[1, 1].set_title("Boxplot dos Graus dos Nós")
    
    # Ajusta layout
    plt.tight_layout()
    plt.show()

    num_strongly_connected = nx.number_strongly_connected_components(G)  # Componentes fortemente conectados
    num_weakly_connected = nx.number_weakly_connected_components(G)
    # Criando o gráfico
    labels = ["Fortemente Conectados", "Fracamente Conectados"]
    values = [num_strongly_connected, num_weakly_connected]

    plt.figure(figsize=(8, 5))
    plt.bar(labels, values, color=["red", "blue"])
    plt.ylabel("Quantidade de Componentes")
    plt.title("Comparação entre Componentes Fortemente e Fracamente Conectados")
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Exibir os valores em cima das barras
    for i, v in enumerate(values):
        plt.text(i, v + 0.05 * max(values), str(v), ha='center', fontsize=11, fontweight='bold')

    plt.show()

    scc_sizes = [len(c) for c in nx.strongly_connected_components(G)]
    
    scc_sizes = np.array([len(c) for c in nx.strongly_connected_components(G)])
    scc_sizes = scc_sizes[scc_sizes > 1]  # Remove SCCs unitários (ruído)

    plt.figure(figsize=(10, 5))
    plt.hist(scc_sizes, bins=np.logspace(0.1, np.log10(max(scc_sizes)), 50), color='red', alpha=0.7)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel("Tamanho do SCC (log)")
    plt.ylabel("Frequência (log)")
    plt.title("Distribuição Log-Log dos SCCs no Grafo de NFTs")
    plt.grid(alpha=0.5)
    plt.show()

    # Detecta SCCs (Componentes Fortemente Conectados)
    sccs = list(nx.strongly_connected_components(G))
    
    # Filtra SCCs pequenos (por exemplo, de tamanho 1 ou 2)
    sccs_filtered = [scc for scc in sccs if len(scc) > 2]
    num_sccs_filtered = len(sccs_filtered)

    # Detecta WCCs (Componentes Fracamente Conectados)
    wccs = list(nx.weakly_connected_components(G))
    num_wccs = len(wccs)

    print(f"Componentes Fortemente Conectados (SCCs) antes do filtro: {len(sccs)}")
    print(f"Componentes Fortemente Conectados (SCCs) após o filtro: {num_sccs_filtered}")
    print(f"Componentes Fracamente Conectados (WCCs): {num_wccs}")

    # Criando um gráfico comparativo
    labels = ["WCCs (Fracamente Conectados)", "SCCs Filtrados (Fortemente Conectados)"]
    values = [num_wccs, num_sccs_filtered]

    plt.figure(figsize=(8, 5))
    plt.bar(labels, values, color=['blue', 'red'])
    plt.ylabel("Quantidade de Componentes")
    plt.title("Comparação SCCs x WCCs (Apenas SCCs Significativos)")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    # Exibir os valores em cima das barras
    for i, v in enumerate(values):
        plt.text(i, v + 0.05 * max(values), str(v), ha='center', fontsize=11, fontweight='bold')
    plt.show()
    

# Exemplo de uso com um grafo direcionado (substituir por seu grafo real)
# G = nx.DiGraph()  # Criar grafo direcionado
analyze_graph_with_visuals(G)
```
